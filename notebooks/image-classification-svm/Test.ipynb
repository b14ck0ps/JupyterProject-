```python
import os
import random
import glob
import pickle
import matplotlib.pyplot as plt
from skimage.transform import resize
from matplotlib import image
```

```python
Categories = ['benign', 'malignant']
model = pickle.load(open('img_model.p', 'rb'))

test_imgs = glob.glob("./test/benign/*.jpg") + \
    glob.glob("./test/malignant/*.jpg")


def test():
    img_path = random.choice(test_imgs)
    folder_name = os.path.basename(os.path.dirname(img_path))

    img = image.imread(img_path)

    img_resize = resize(img, (150, 150, 3))
    l = [img_resize.flatten()]
    probability = model.predict_proba(l)

    BENIGN = probability[0][0]*100
    MALIGNANT = probability[0][1]*100

    prediction = Categories[model.predict(l)[0]]
    prediction_accuracy = False
    print("\nExpected → " + folder_name + " || Predicted → " + prediction)

    if prediction == folder_name:
        print("\nCorrect Prediction".upper() + "✅")
        accuracy = BENIGN if prediction == "benign" else MALIGNANT
        prediction_accuracy = True
        print("Accuracy: {:.2f}%".format(accuracy))

    else:
        print("\nWrong Prediction".upper()+ "❌")
        print("Benign: {:.2f}%".format(BENIGN))
        print("Malignant: {:.2f}%".format(MALIGNANT))
    print("Data Link: "+img_path)

    # plt.figure(dpi=30)
    # plt.imshow(img)
    # plt.show()
    return prediction_accuracy


def test_accuracy(n=1):
    saperator = '-'*60
    correct = 0
    for i in range(n):
        print("Test {}".format(i+1))
        if test():
            correct += 1
        print("\n"+saperator+"\n")
        
    print("Total Tests: {}".format(n))
    print("Total Accuracy: {:.2f}%".format(correct/n*100))
```

```python
test_accuracy(500)
```
